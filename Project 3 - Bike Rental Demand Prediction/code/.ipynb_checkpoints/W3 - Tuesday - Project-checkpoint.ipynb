# 4: Adding date time feature and remapping Season Column

```python
df = bike_train_df.copy()
df.reset_index(inplace=True)
```

```python
#creating a three new columns as below to split any given day:
# daytime basically to split the day between morning, afternoon, evening and midnight
# The splitting is done by looking at the bike demand graph in a given day
# daytimetag is counting from 1 to 4 with 1 being morning and 4 being midnight
# daytimetag1 is just hour hand of the day
daytime= []
daytimetag = []
daytimetag1 = []
for i in df["datetime"]:
    if i.hour >=6 and i.hour < 11:
        daytime.append("morning")
        daytimetag.append(1)
        daytimetag1.append(i.hour)
    elif i.hour >= 11 and i.hour < 15:
        daytime.append("afternoon")
        daytimetag.append(2)
        daytimetag1.append(i.hour)
    elif i.hour >= 15 and i.hour <= 20:
        daytime.append("evening")
        daytimetag.append(3)
        daytimetag1.append(i.hour)
    else:
        daytime.append("midnight")
        daytimetag.append(4)
        daytimetag1.append(i.hour)
```

```python
# March to May is spring and is given season code 1
# June to August is Summer and is given season code 2
# September to November is Fall and is given season code 3
# December to February is Winter and is given season code 4
season_tag = []
season_code = []
season_code1 = []
for i in df["datetime"]:
    if i.month >=3 and i.month < 6:
        season_tag.append("spring")
        season_code.append(1)
        season_code1.append(i.month)
    elif i.month >= 6 and i.month < 9:
        season_tag.append("summer")
        season_code.append(2)
        season_code1.append(i.month)
    elif i.hour >= 9 and i.month < 12:
        season_tag.append("fall")
        season_code.append(3)
        season_code1.append(i.month)
    else:
        season_tag.append("winter")
        season_code.append(4)
        season_code1.append(i.month)
```

```python
bike_train_df["daytime"] = daytime
bike_train_df["daytimetag"] = daytimetag
bike_train_df["daytimetag1"] = daytimetag1
bike_train_df["season_map"] = season_tag
bike_train_df["season_tag_corr"] = season_code
bike_train_df["season_tag_corr1"] = season_code1
```

# 6: Feature Engineering

#### 6.1 One hot Encoding of season, weather, daytimetag
* **Observation:** Significant improvement in train/ test results
* **Conclusion:** Use one hot encoding of **season, weather and daytimetag**

```python
#Check tomorrow whats happening????
X = bike_train_df[["atemp","humidity","season","weather","workingday","windspeed","holiday","daytimetag"]]
y = bike_train_df["count"]

Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.20,random_state=141)

#One hot encoding weather, season and daytimetag
Xtrain[["W_1","W_2","W_3"]] = pd.get_dummies(Xtrain['weather'])
Xtrain[["S_1","S_2","S_3","S_4"]] = pd.get_dummies(Xtrain['season'])
Xtrain[["D_1","D_2","D_3","D_4"]] = pd.get_dummies(Xtrain['daytimetag'])
Xtrain = Xtrain.drop(["W_3","S_1","season","weather","D_1","daytimetag"
                     ],axis=1)


Xtest[["W_1","W_2","W_3","W_4"]] = pd.get_dummies(Xtest['weather'])
Xtest[["S_1","S_2","S_3","S_4"]] = pd.get_dummies(Xtest['season'])
Xtest[["D_1","D_2","D_3","D_4"]] = pd.get_dummies(Xtest['daytimetag'])
Xtest = Xtest.drop(["W_3","W_4","S_1","season","weather","D_1","daytimetag"
                   ],axis=1)

```

```python
model = LinearRegression()
model.fit(Xtrain,ytrain)
modelfe_train_score = model.score(Xtrain,ytrain)
modelfe_test_score = model.score(Xtest,ytest)
f' train: {modelfe_train_score}, test: {modelfe_test_score}'
```

#### 6.2 Polynomial Feature Expansion
* **Observation:** Significant improvement in train/ test results
* **Conclusion:** Use Polynomial feature expansion

```python
pl = PolynomialFeatures(degree=2, include_bias=True)
m = LinearRegression()
pipeline = Pipeline([("polynomial_features", pl),
                         ("linear_regression", m)])
pipeline.fit(Xtrain,ytrain)
pipeline.fit(Xtest,ytest)
pipeline.score(Xtrain,ytrain), pipeline.score(Xtest,ytest)
```

#### How to use column transformer?

```python
#how to use column transformer to only transform atemp humidity and windspeed?
pl = PolynomialFeatures(degree=2, include_bias=True)
transform_select = ColumnTransformer("polynomial_features",pl,["atemp","humidity","windspeed"])
m = LinearRegression()
pipeline = Pipeline([("polynomial_features", transform_select),
                         ("linear_regression", m)])
pipeline.fit(Xtrain,ytrain)
pipeline.fit(Xtest,ytest)
pipeline.score(Xtrain,ytrain), pipeline.score(Xtest,ytest)
```
